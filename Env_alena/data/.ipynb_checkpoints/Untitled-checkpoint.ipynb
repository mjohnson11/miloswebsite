```python
import pandas as pd
from matplotlib import pyplot as pl
import seaborn as sns
import numpy as np
from glob import glob

%matplotlib inline
```

```python
barcode_reads = pd.read_csv('Filtered_strains_reads.csv')
```

```python
barcode_reads['Experiment'] = barcode_reads['Strain']+'_'+barcode_reads['Environment']
barcode_reads['Exp_rep'] = barcode_reads['Experiment']+'_R'+barcode_reads['Bio_rep'].map(str)
barcode_reads['MutID_BC'] = barcode_reads['Mut_ID'].map(str)+'_'+barcode_reads['BC_ID'].map(str)
```

```python
info = {str(i[0]): i[1] for i in np.array(pd.read_csv('Mut_full_info.csv')[['Mut_ID', 'Gene.Use']])}
edge_to_num = {i[0]:i[1] for i in np.array(pd.read_csv('Mut_full_info.csv')[['Edge', 'Mut_ID']])}
edge_to_gene = {i[0]:i[1] for i in np.array(pd.read_csv('Mut_full_info.csv')[['Edge', 'Gene.Use']])}
```

```python
for er in set(barcode_reads['Exp_rep']):
    tmp = barcode_reads[barcode_reads['Exp_rep']==er]
    if len(set(tmp['Flask'])) > 1: # For LK5-G01 which is in multiple flasks, we'll use flask 1 data
        tmp = tmp[tmp['Flask']==1]
    one_rep = pd.DataFrame(tmp[['MutID_BC', 'Time', 'Reads']].pivot(index='MutID_BC', columns='Time', values='Reads'))
    one_rep['Mut_ID'] = one_rep.index.str.split('_').str[0]
    one_rep['BC_ID'] = one_rep.index.str.split('_').str[1]
    one_rep['Gene'] = one_rep['Mut_ID'].map(info)
    one_rep.iloc[:3]
    tots = [np.nansum(one_rep['T'+str(i)]) for i in range(5)]
    for i in range(5):
        one_rep['F'+str(i)] = np.log10(one_rep['T'+str(i)]/tots[i])
    one_rep['Log10_freq_trajectory'] = one_rep.apply(lambda row: ';'.join([str(row['F'+str(i)]) for i in range(5)]), axis=1)
    one_rep[['Mut_ID', 'BC_ID', 'Gene', 'T0', 'T1', 'T2', 'T3', 'T3', 'Log10_freq_trajectory']].to_csv('BC_freqs/'+er+'.csv', index=False)
```

## BC trajectories for my data:

```python
segs = list(set(barcode_reads['Strain']))
len(segs), segs[:2]
```

```python
for seg in segs:
    for r in range(1,3):
        try:
            tmp = pd.read_csv('../../figure80/stories/interaction_action/data/freq_files_new/' + seg + '_rep' + str(r) + '_logfreqs.csv')
            tmp['Mut_ID'] = tmp['Edge'].map(edge_to_num)
            tmp['Gene'] = tmp['Mut_ID'].map(edge_to_gene)
            tmp['BC_ID'] = tmp['cBC']
            for i in range(5):
                cols = [c for c in tmp if c.split('-')[-1]=='T'+str(i)]
                if len(cols) == 1:
                    tmp['T'+str(i)] = tmp[cols[0]]
                else:
                    tmp['T'+str(i)] = [np.nan]*len(tmp)
            tots = [np.nansum(tmp['T'+str(i)]) for i in range(5)]
            for i in range(5):
                tmp['F'+str(i)] = np.log10(tmp['T'+str(i)]/tots[i])
            tmp['Log10_freq_trajectory'] = tmp.apply(lambda row: ';'.join([str(row['F'+str(i)]) for i in range(5)]), axis=1)
            tmp[['Mut_ID', 'BC_ID', 'Gene', 'T0', 'T1', 'T2', 'T3', 'T3', 'Log10_freq_trajectory']].to_csv('BC_freqs/'+seg+'_YPD_R'+str(r)+'.csv', index=False)
        except FileNotFoundError:
            print('No file:', seg + '_rep' + str(r))
```

```python
muts = pd.read_csv('SC_Mut.csv')
flist = glob('../../figure80/stories/interaction_action/data/edge_files/*.csv')
seg_fits = dict()
for f in flist[:1]:
    tmp = pd.read_csv(f).rename(columns={'Segregant': 'Strain', 's': 'Sel_coef'})
    tmp = tmp[tmp['Strain'].isin(segs)]
    seg_fits.update({i[0]:i[1] for i in np.array(tmp[['Strain', 'Seg.Fitness']])})
    tmp['Bio_rep'] = [1]*len(tmp)
    tmp['Flask'] = [1]*len(tmp)
    tmp['Mut_ID'] = [edge_to_num[f.split('/')[-1].split('_')[0]]]*len(tmp)
    tmp['Environment'] = ['YPD']*len(tmp)
    muts = pd.concat([muts, tmp[['Environment', 'Bio_rep', 'Flask', 'Strain', 'Mut_ID', 'Sel_coef']]])

muts['Experiment'] = muts['Strain']+'_'+muts['Environment']
muts['Experiment_Mut'] = muts['Experiment']+'_Mut'+muts['Mut_ID'].apply(lambda m: str(m))
print(len(muts), len(set(muts['Experiment'])), len(set(muts['Experiment_Mut'])))
muts = muts[['Experiment', 'Strain', 'Environment', 'Experiment_Mut', 'Sel_coef']].groupby(['Experiment', 'Strain', 'Environment','Experiment_Mut'], as_index=False).mean()
muts.iloc[:3]
```

```python
td = pd.read_csv('Avg_GR_WT_strains.csv')
td['Experiment'] = td['Strain']+'_'+td['Environment']
td = td[['Strain', 'Environment', 'Experiment', 'Flask', 'Avg_GR']].groupby(['Strain', 'Environment', 'Experiment', 'Flask'], as_index=False).mean()
mat = []
for s in seg_fits:
    mat.append([s, 'YPD', s+'_YPD', 1, seg_fits[s]+0.3]) # ARBITRARY +0.3 converting fitness to GR
td = pd.concat([td, pd.DataFrame(mat, columns=['Strain', 'Environment', 'Experiment', 'Flask', 'Avg_GR'])])
td.iloc[-3:]
```

```python
dfe_mean = muts[['Experiment', 'Sel_coef']].groupby('Experiment', as_index=False).mean().rename(columns={'Sel_coef': 'DFE_mean'})
td = td.merge(dfe_mean, on='Experiment', how='inner')
envs = ['30_SC3', '30_SC5', '30_SC7', '37_SC3', '37_SC5', '37_SC7'] # Not including YPD in stats (max and std)
envs_all = ['YPD', '30_SC3', '30_SC5', '30_SC7', '37_SC3', '37_SC5', '37_SC7'] # Not including YPD in these things
td['Env_GR'] = td['Strain'].apply(lambda s: ';'.join([str(td[td['Experiment']==s+'_'+env].iloc[0]['Avg_GR']) for env in envs_all]))
td['Env_DFE_mean'] = td['Strain'].apply(lambda s: ';'.join([str(td[td['Experiment']==s+'_'+env].iloc[0]['DFE_mean']) for env in envs_all]))
td['GR_std'] = td['Strain'].apply(lambda s: np.nanstd([td[td['Experiment']==s+'_'+env].iloc[0]['Avg_GR'] for env in envs]))
td['GR_max'] = td['Strain'].apply(lambda s: np.nanmax([td[td['Experiment']==s+'_'+env].iloc[0]['Avg_GR'] for env in envs]))
td['GR_std_max_excluded'] = td.apply(lambda row: np.nanstd([td[td['Experiment']==row['Strain']+'_'+env].iloc[0]['Avg_GR'] for env in envs 
                                                            if td[td['Experiment']==row['Strain']+'_'+env].iloc[0]['Avg_GR']!=row['GR_max']]), axis=1)
td['Env_series'] = [';'.join(envs_all)]*len(td)
td.to_csv('GR_v_DFE_mean.tsv', sep='\t', index=False)
td.iloc[:3]
```

```python
muts['Mut'] = muts['Experiment_Mut'].str.split('_').str[-1]
muts['Gene'] = muts['Mut'].apply(lambda m: info[m[3:]])
muts = muts.merge(td[['Experiment', 'Avg_GR']], on='Experiment', how='left')
muts.iloc[:3]
```

```python
muts.to_csv('Mut_s_formatted.tsv', sep='\t', index=False)
```
